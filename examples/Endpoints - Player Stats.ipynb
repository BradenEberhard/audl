### How to get Player Stats

```python
from audl.stats.endpoints.playerstats import PlayerStats

# from https://theaudl.com/stats/player-stats
total = PlayerStats('career', 'total', 'breeze').fetch_table(show_message=False)  
games = PlayerStats('career', 'game', 'breeze').fetch_table()  
points = PlayerStats('career', 'points', 'breeze').fetch_table()  
possessions = PlayerStats('career', 'possessions', 'breeze').fetch_table()  
minutes = PlayerStats('career', 'minutes', 'breeze').fetch_table()
```
